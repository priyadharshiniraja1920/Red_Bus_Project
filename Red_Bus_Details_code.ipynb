```python
from selenium import webdriver
from selenium.webdriver import ActionChains
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
from selenium.common.exceptions import TimeoutException, NoSuchElementException
import pandas as pd
```

## 1. Andhra Pradesh Bus Details

```python
read_AP_df = pd.read_csv(r"C:/Users/PRIYA/OneDrive/Desktop/Red_Bus_Project/ap_df.csv")
read_AP_df
```

```python
AP_driver = webdriver.Chrome()
AP_Bus_names = []
AP_Bus_types = []
AP_Start_Time = []
AP_End_Time = []
AP_Ratings = []
AP_Total_Duration = []
AP_Prices = []
AP_Seats_Available = []
AP_Route_names = []
AP_Route_links = []

for i,r in read_AP_df.iterrows():
    link=r["Route_Links"]
    routes=r["Route_Names"]
    AP_driver.get(link)
    time.sleep(2)  

    elements = AP_driver.find_elements(By.XPATH, f"//a[contains(@href, '{link}')]")
    for element in elements:
        element.click()
        time.sleep(2)

    try:
        clicks = AP_driver.find_element(By.XPATH, "//div[@class='button']")
        clicks.click()
    except:
        continue  
    time.sleep(2)
    
    scrolling = True
    while scrolling:
        old_page_source = AP_driver.page_source
        ActionChains(AP_driver).send_keys(Keys.PAGE_DOWN).perform()
        time.sleep(5)
        new_page_source = AP_driver.page_source
        if new_page_source == old_page_source:
            scrolling = False

    bus_name = AP_driver.find_elements(By.XPATH, "//div[@class='travels lh-24 f-bold d-color']")
    bus_type = AP_driver.find_elements(By.XPATH, "//div[@class='bus-type f-12 m-top-16 l-color evBus']")
    start_time = AP_driver.find_elements(By.XPATH, "//*[@class='dp-time f-19 d-color f-bold']")
    end_time = AP_driver.find_elements(By.XPATH, "//*[@class='bp-time f-19 d-color disp-Inline']")
    total_duration = AP_driver.find_elements(By.XPATH, "//*[@class='dur l-color lh-24']")
    try:
        rating = AP_driver.find_elements(By.XPATH,"//div[@class='clearfix row-one']/div[@class='column-six p-right-10 w-10 fl']")
    except:
        continue
    price = AP_driver.find_elements(By.XPATH, '//*[@class="fare d-block"]')
    seats = AP_driver.find_elements(By.XPATH, "//div[contains(@class, 'seat-left')]")

    for bus in bus_name:
        AP_Bus_names.append(bus.text)
        AP_Route_links.append(link)
        AP_Route_names.append(routes)
    for bus_type_elem in bus_type:
        AP_Bus_types.append(bus_type_elem.text)
    for start_time_elem in start_time:
        AP_Start_Time.append(start_time_elem.text)
    for end_time_elem in end_time:
        AP_End_Time.append(end_time_elem.text)
    for total_duration_elem in total_duration:
        AP_Total_Duration.append(total_duration_elem.text)
    for ratings in rating:
        AP_Ratings.append(ratings.text)
    for price_elem in price:
        AP_Prices.append(price_elem.text)
    for seats_elem in seats:
        AP_Seats_Available.append(seats_elem.text)
print("Successfully Completed")
```

```python
data_1 = {
    'Bus_name': AP_Bus_names,
    'Bus_type': AP_Bus_types,
    'Start_time': AP_Start_Time,
    'End_time': AP_End_Time,
    'Total_duration': AP_Total_Duration,
    'Price': AP_Prices,
    "Seats_Available": AP_Seats_Available,
    "Ratings":AP_Ratings,
    'Route_Links': AP_Route_links,
    'Route_Names': AP_Route_names
}

Andhra_Details_df = pd.DataFrame(data_1)
path=r"C:/Users/PRIYA/OneDrive/Desktop/Red_Bus_Project/Andhra_Details_df.csv"
Andhra_Details_df.to_csv(path,index=False)
```
